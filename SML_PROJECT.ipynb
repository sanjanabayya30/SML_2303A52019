# LOGISTIC REGRESSION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix
```

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/archive.zip')
```

```python
# Define target variable
target = 'Sleep_Quality_Score'
```

```python
# Define features
features = ['Heart_Rate_Variability', 'Body_Temperature', 'Bedtime_Consistency']
```

```python
# Convert target variable into numerical values (0/1) using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[target] = le.fit_transform(df[target])
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
# Create logistic regression model
model = LogisticRegression(max_iter=1000)
```

```python
# Train model
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
conf_mat = confusion_matrix(y_test, y_pred)
```

```python
# Print metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("F1 Score:", f1)
print("Recall:", recall)
print("Confusion Matrix:\n", conf_mat)
```
