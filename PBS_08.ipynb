```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr

# Data
data = {
    'Daily Calorie Intake': [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000],
    'Exercise Duration': [60, 50, 30, 90, 45, 20, 75, 95, 15, 10],
    'BMI': [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]
}

df = pd.DataFrame(data)

# 1. Scatter Plots
plt.figure(figsize=(15, 5))

# Daily Calorie Intake vs. BMI
plt.subplot(1, 3, 1)
plt.scatter(df['Daily Calorie Intake'], df['BMI'])
plt.title('Daily Calorie Intake vs. BMI')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')

# Exercise Duration vs. BMI
plt.subplot(1, 3, 2)
plt.scatter(df['Exercise Duration'], df['BMI'])
plt.title('Exercise Duration vs. BMI')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('BMI')

# Daily Calorie Intake vs. Exercise Duration
plt.subplot(1, 3, 3)
plt.scatter(df['Daily Calorie Intake'], df['Exercise Duration'])
plt.title('Daily Calorie Intake vs. Exercise Duration')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration (minutes)')

plt.tight_layout()
plt.show()

# 2. Pearson Correlation Coefficients
# Daily Calorie Intake vs. BMI
r1, p1 = pearsonr(df['Daily Calorie Intake'], df['BMI'])
print(f"Pearson correlation (Calories vs BMI): {r1:.3f}")

# Exercise Duration vs. BMI
r2, p2 = pearsonr(df['Exercise Duration'], df['BMI'])
print(f"Pearson correlation (Exercise vs BMI): {r2:.3f}")

# Daily Calorie Intake vs. Exercise Duration
r3, p3 = pearsonr(df['Daily Calorie Intake'], df['Exercise Duration'])
print(f"Pearson correlation (Calories vs Exercise): {r3:.3f}")

# 3. Interpretation
print("\nInterpretation:")
print("- Daily Calorie Intake vs BMI shows a positive correlation (r = {:.3f}), suggesting higher calorie intake is associated with higher BMI.".format(r1))
print("- Exercise Duration vs BMI shows a strong negative correlation (r = {:.3f}), indicating more exercise is associated with lower BMI.".format(r2))
print("- Daily Calorie Intake vs Exercise Duration shows a weak negative correlation (r = {:.3f}), suggesting people who consume more calories tend to exercise slightly less.".format(r3))
```
