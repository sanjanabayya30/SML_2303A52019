```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the seed for reproducibility
np.random.seed(0)
# Generate a normal distribution with mean 60 and standard deviation 12
mean = 60
std_dev = 12
data = np.random.normal(mean, std_dev, 1000)
# Plot a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
# Analyze the symmetry and bell-shaped curve
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
# Visualize the Probability Density Function (PDF)
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = np.exp(-((x - mean) / std_dev)**2 / 2) / (std_dev * np.sqrt(2 * np.pi))
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='red')
plt.title("Probability Density Function (PDF) of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Mean and standard deviation of daily calorie intake
mean = 30
std_dev = 5
# Define the values for the probabilities to be calculated
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
# Calculate the probabilities
prob_greater_equal_19 = 1 - stats.norm.cdf(19, mean, std_dev)
prob_less_equal_19 = stats.norm.cdf(19, mean, std_dev)
prob_between_19_and_35 = stats.norm.cdf(35, mean, std_dev) - stats.norm.cdf(19, mean, std_dev)
# Print the probabilities
print(f'Probability of intake greater than or equal to 19 calories: {prob_greater_equal_19:.4f}')
print(f'Probability of intake less than or equal to 19 calories: {prob_less_equal_19:.4f}')
print(f'Probability of intake between 19 and 35 calories: {prob_between_19_and_35:.4f}')
# Plot the normal distribution and shade the areas
plt.figure(figsize=(10, 6))
plt.plot(x, stats.norm.pdf(x, mean, std_dev), label='Normal Distribution')
# Shade area for P(X >= 19)
x_fill_greater = np.linspace(19, mean + 3*std_dev, 1000)
plt.fill_between(x_fill_greater, stats.norm.pdf(x_fill_greater, mean, std_dev), alpha=0.3, label='X >= 19')
# Shade area for P(X <= 19)
x_fill_less = np.linspace(mean - 3*std_dev, 19, 1000)
plt.fill_between(x_fill_less, stats.norm.pdf(x_fill_less, mean, std_dev), alpha=0.3, label='X <= 19')
# Shade area for P(19 <= X <= 35)
x_fill_between = np.linspace(19, 35, 1000)
plt.fill_between(x_fill_between, stats.norm.pdf(x_fill_between, mean, std_dev), alpha=0.3, label='19 <= X <= 35')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Daily Calorie Intake')
plt.legend()
plt.show()
```
